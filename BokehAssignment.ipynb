```python
Q1 ->How can you create a Bokeh plot using Python code?


Ans -> 
from bokeh.plotting import figure , show , output_file

x = [1,2,3,4,5]
y = [6,7,2,4,5]

output_file("plot.html")

p = figure(title = "sample bokeh")
p.xaxis.axis_label =  "x axis"
p.yaxis.axis_label =  "y axis"
p.line(x,y , legend_label = "Line" , line_width = 2)
p.circle(x , y , legend_label = "circle" , fill_color = "red" , size = 8)
show(p)
```

```python
Q2 ->  What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

Ans -> 
from bokeh.plotting import figure , output_file , show

x = [1,2,3,4,5]
y = [6,7,2,4,5]

output_file("gylphs.html")
p = figure(title = "Bokeh with gylph")
p.xaxis.axis_label = "x axis"
p.yaxis.axis_label = "y axis"
p.circle(x , y , color = "blue" , alpha = 0.5)
p.square(x,y , color = "black" , alpha = 1)
show(p)
```

```python
Q3 -> How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

Ans -> 
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

from bokeh.plotting import figure , show , output_file
p = figure(title = "My plot")

x = [1,2,3,4,5]
y = [2,3,4,5,6]
p.xaxis.axis_label = "x axis"
p.xaxis.axis_line_width = 2
p.yaxis.axis_label = "y axis"
p.yaxis.axis_line_width = 2
p.yaxis.axis_line_color = "red"

p.circle(x, y ,legend = "Data")
show(p)
p.legend.location = "top_left"
```

```python
Q4 -> What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?

Ans -> 
from bokeh.plotting import figure , show , output_file
plot = figure()
plot.circle(x,y)
  
from bokeh.events import ButtonClick
def update_plot():
  plot.on_event(ButtonClick , update_plot)
```

```python
Q5 -> How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

Ans -> 

from flask import Flask , render_template
from bokeh.embed import components
from bokeh.plotting import figure
from bokeh.resources import CDN

app = Flask(__name__)

@app.route('/')
def home():
  plot = figure()
  plot.circle([1,2,3] , [4,5,6])

  script , div = components(plot , CDN)

  return render_template('home.html' , script = script , div = div)

if __name__ == '__main__':
  app.run(debug = True)



<!DOCTYPE html>
<html>
<head>
 <title> My Bokeh plot </title>
    {{ script|safe }}
  
</head>
<body>
    {{ div|safe }}
</body>
</html>
```
